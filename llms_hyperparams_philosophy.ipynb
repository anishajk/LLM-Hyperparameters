```python
# YOUR CODE HERE
# import the necessary packages
# include appropriate doc strings for functions
# add more cells as needed/desired

import numpy as np
import matplotlib.pyplot as plt

words = ["in", "school", "sky", "pretty", "moon", "awesome", "kind", "rain", "fall", "start"]

z = np.random.uniform(-1,1,len(words))

def softmax(z, temp = 1):
    ez = np.exp(z/temp)
    sum_ez = np.sum(ez)
    sftmz = ez/sum_ez
    return sftmz 


sftmx = softmax(z)
print("Softmax values for each word: ",sftmx)
    
temperature = [0.23, 1, 1.85]

for t in temperature:
    sm_z_t = softmax(z,t)
    plt.plot(sm_z_t, label = t)
    
# Plotting graph for Softmax Values with different temperatures
plt.xlabel('Vocabulary')
plt.ylabel('Softmax values')
plt.title('Softmax Values with different temperatures ')   

plt.legend()
plt.show()
```

```python
# C. Changing the top-k parameter
import random
def topKSampling(k):
    topIndices = np.argsort(sftmx)[::-1][:k]
    selectedWords = [words[i] for i in topIndices]
    selectedProb = [sftmx[i] for i in topIndices]
    sumTopK = sum(selectedProb)
    for j in range(len(selectedProb)):        
        selectedProb[j] /= sumTopK        
    choosenWord = random.choices(selectedWords, weights=selectedProb)
    print("\n","Top K value: ", k)
    print("Top K Words: ",selectedWords )
    print("Probability of Top K Words: ",selectedProb)
    print("Selceted Word: ", choosenWord)
    
topK = [1,2,3,5]
for k in topK:
    topKSampling(k)


```
